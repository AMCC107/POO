```python
def maximovalor(A):
    count_for=0
    count_if=0 
    maxval = A[0]  # Inicializar maxval con el primer elemento de la lista
    for i in range(len(A)):
        count_for+=1
        if A[i] > maxval:
            count_if+=1
            maxval = A[i]  # Actualizar maxval si el elemento actual es mayor
    print(f"Iteraciones del ciclo for: {count_for}")
    print(f"Condición if evaluada como verdadera: {count_if}")
    return maxval
B = [0, 1, 2, 3, 4, 5]
print(maximovalor(B))  # Esto imprimirá 5, que es el valor máximo
```

```python
def same_value(A):
    count_outer = 0  # Contador para el ciclo externo
    count_inner = 0  # Contador para el ciclo interno

    for i in range(len(A) - 1):
        print(count_outer)
        count_outer += 1  # Incrementa el contador del ciclo externo
        print(count_outer)
        for j in range(i + 1, len(A)):
            count_inner += 1  # Incrementa el contador del ciclo interno

            if A[i] == A[j]:
                print(f"  Se encontró un valor duplicado: {A[i]}")
                return False

    print(f"Total de iteraciones del ciclo externo: {count_outer}")
    print(f"Total de iteraciones del ciclo interno: {count_inner}")
    return True

B = [1, 2, 3, 4,5]
print(same_value(B))
```

```python
def matriz_recorrido(A, B):
    # Inicializamos C con ceros
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]

    for i in range(n):  # Iterar sobre las filas de A
        for j in range(n):  # Iterar sobre las columnas de B
            C[i][j] = 0  # Inicializar el elemento de la posición (i, j)
            for k in range(n):  # Iterar sobre los elementos de la fila y columna
                C[i][j] += A[i][k] * B[k][j]  # Producto escalar y suma acumulada
    return C
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Llamada a la función
print(matriz_recorrido(A, B))
```

```python
def matriz_recorrido(A, B):
    # Inicializamos C con ceros
    n = len(A)
    C = [[0 for _ in range(n)] for _ in range(n)]

    # Contadores de iteraciones
    count_outer = 0  # Iteraciones del ciclo externo
    count_middle = 0  # Iteraciones del ciclo medio
    count_inner = 0  # Iteraciones del ciclo interno

    for i in range(n):  # Iterar sobre las filas de A
        count_outer += 1
        print(f"Ciclo externo, iteración {count_outer}: i = {i}")

        for j in range(n):  # Iterar sobre las columnas de B
            count_middle += 1
            print(f"  Ciclo medio, iteración {count_middle}: j = {j}")

            C[i][j] = 0  # Inicializar el elemento de la posición (i, j)

            for k in range(n):  # Iterar sobre los elementos de la fila y columna
                count_inner += 1
                print(f"    Ciclo interno, iteración {count_inner}: k = {k}")
                C[i][j] += A[i][k] * B[k][j]  # Producto escalar y suma acumulada
                print(f"    Actualizando C[{i}][{j}]: {C[i][j]} (A[{i}][{k}] = {A[i][k]}, B[{k}][{j}] = {B[k][j]})")

    # Imprimir totales de iteraciones
    print(f"Total de iteraciones del ciclo externo: {count_outer}")
    print(f"Total de iteraciones del ciclo medio: {count_middle}")
    print(f"Total de iteraciones del ciclo interno: {count_inner}")
    return C

# Ejemplo de matrices A y B
A = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
B = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Llamada a la función
print(matriz_recorrido(A, B))
```
